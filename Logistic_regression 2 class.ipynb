```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#cost function
def cost(x,y,theta,b):
    m=len(y)#no of training set
    J=0
    for i in range(m):
        h=1/(1+np.e**(-np.dot(theta,np.transpose(x[i])))+b)
        J+=-(1/m)*((y[i]*np.log(h))+(1-y[i])*np.log(1-h))
        
    return J
    
    
#derivatives 
def derivative(x,y,theta,b):
    m=len(y)#no of training examples
    n=len(theta)#no of features
    d=np.zeros(n)
    db=0.
    for j in range(m):
        h=1/(1+np.e**(-np.dot(theta,np.transpose(x[j]))))
        for i in range(n):
            d[i]+=(1/m)*(h-y[j])*x[j][i]
        db+=(1/m)*(h-y[j])
    return d,db

```

```python
dataset=pd.read_csv("ex2data1.csv")
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
y=y.reshape(len(y),1)
m,n=x.shape
```

```python
theta_ini=np.zeros(n)
b=0.
alpha=1e-4
t1=[]
t=[]
temp=np.zeros(n)
```

```python
#main loop

for it in range(10000):
    d,db=derivative(x,y,theta_ini,b)
    
    theta_ini=theta_ini-alpha*d
    b=b-alpha*db
    
    t.append(cost(x,y,theta_ini,b))
    print(cost(x,y,theta_ini,b))
    t1.append(theta_ini[0])
    
```
